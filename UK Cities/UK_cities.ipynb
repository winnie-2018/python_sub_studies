```python
info = df['Population'].describe()

mean = int(np.round(info['mean']))
q1 = int(np.round(info['25%']))
q3 = int(np.round(info['75%']))
min = int(np.round(info['min']))
max = int(np.round(info['max']))

print(q1, mean, q3, min, max)
```

```python
coords = []
long=[]
lat=[]

for index in df['Longitude']:
    long.append(index)

for index in df['Latitude']:
    lat.append(index)

for index in range(len(df['Latitude'])):
    coord = [lat[index], long[index]]
    coords.append(coord)
```

Declare the coordinate

```python
# Use Folium
m = folium.Map(location=(51.5074, -0.1278), zoom_control=True, zoom_start=6)

for coord in coords:
    i = coords.index(coord)
    pop = df.iloc[i]['Population']
    city = df.iloc[i]['City']
    
    stk = False

    if pop == max:
        c = 'red',
        stk = True
    elif pop == min:
        c = 'aquamarine',
        stk = True
    elif pop <= q1:
        c = 'cyan'
    elif pop <= mean and pop > q1:
        c = 'orange'
    elif pop <= q3 and pop > mean:
        c = 'purple'
    elif pop > q3: 
        c = 'salmon'
    else: 
        c = 'grey'
        
    folium.CircleMarker(
        location=coord,
        radius=25,
        color=c,
        stroke=stk,
        fill=True,
        fill_opacity=0.5,
        opacity=1,
        popup=folium.Popup(f'{city} Population: {str(pop)}'),
    ).add_to(m)
    
m.save("uk_cities_population.html")
```
